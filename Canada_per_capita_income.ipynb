```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
# Create a DataFrame for missing years (2017 to 2030)
new_years = pd.DataFrame({'year': range(2017, 2031)})
new_predictions = reg.predict(new_years)
```

```python
# Create a DataFrame for the predicted values
new_data = pd.DataFrame({
    'year': new_years['year'],
    'per capita income (US$)': new_predictions
})
```

```python
# Append the new data to the original DataFrame
df = pd.concat([df, new_data], ignore_index=True)
```

```python
df.tail()
```

```python
plt.scatter(df.year, df['per capita income (US$)'], marker= '+', color= 'red')
plt.plot(df.year, reg.predict(df[['year']]), color= 'blue')
```
